```python
import numpy as np
s=df['persons'].value_counts().keys().tolist()
for i in range(0,len(s)):    
    df[s[i]]=np.where(df['persons']==s[i],1,0)
df=df.drop('persons',axis=1)
df
```

The goal is to remove the column person and replace it wwith n number of columns where n is the number of diff values inside person.This conversion of row values to column values is called as pivoting.
First all the values are stored in a list using the value_counts function on the column person.
Then using numpy array we assign a 1 if the key value matches with the content of the persons column  and if not the value 0 is assigned hence.
Since there are n values we use the len(list) in a loop.
Once n number of columns are created we drop the person column using drop function.

```python
import numpy as np
s=df['safety'].value_counts().keys().tolist()
for i in range(0,len(s)):
    df['safety']=df['safety'].replace(to_replace=s[i],value=i+1)
df   
```

The goal is to convert the values of safety to  numerical values.
First we find the number of values and the values using value_counts on the safety column and store it in a list.
Once we have all the values we use a for loop which goes till len(list) and then compare the values of the list with the values of the column if matched then a new value is assigned to them as i+1.
The replacement is done using the function replace(to_replace,new value).
