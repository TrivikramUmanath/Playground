```python
youthmeter=[]
w=df.Age.mean()
for i in range(0,df.shape[0]):
    if(float(df.iloc[i,0])>w):
        youthmeter.append('Old')
    elif(float(df.iloc[i,0])<w):
        youthmeter.append('Young')
df['youth meter']=youthmeter              
```

```python
def func1(data,mean):
    if(data>mean):
        return 'Old'
    else:
        return 'Young'

df['youth meter']=df['Age'].apply(func1,mean=df.Age.mean())
df
```

So the goal is to add a new column called the youth meter which should have the value Young if the Age is less than the mean age else old.We are using a function func1 here who has two parameters Age and mean.Here the mean age is calculated and assigned to the parameter mean.The mean is then compared with the data i.e in this case the df['Age'].The apply function does the iteration itself and we don't need to use a for loop exclusively for this.

```python
df['youth meter']=df['Age'].apply(lambda val:'Young' if val<df.Age.mean() else 'Old')
df
```

```python
import math
mean=df['Age'].mean()
var=df['Age'].var()
std=math.sqrt(var)
df['z-score']=df.apply(lambda val:(val['Age']-mean)/std,axis=1)
f=(df.loc[(df['Age']<mean+std) & (df['Age']>mean-std)]).shape[0]
s=(df.loc[(df['Age']<mean+2*std) & (df['Age']>mean-2*std)]).shape[0]
t=(df.loc[(df['Age']<mean+3*std) & (df['Age']>mean-3*std)]).shape[0]
print(f*100/df.shape[0])
print(s*100/df.shape[0])
print(t*100/df.shape[0])
if(float(f)==df.shape[0]*0.65 and float(s)==df.shape[0]*0.95 and float(t)==df.shape[0]*0.99):
    print("Normal")
else:
    print("Abnormal")
```

In the above code we are trying to find whether the age attribute is following the normal distribution or not,the above conditions if satisfied then it's normal 65 or close to 65 percent,95 and 99 similarly number of values should be satisfied for a perfect bell shaped graph.
