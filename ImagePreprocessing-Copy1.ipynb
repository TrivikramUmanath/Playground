```python
import cv2
import os
import glob
img_dir = "C:\\Users\\Trivikram\\fanta"  
data_path = os.path.join(img_dir,'*g')
files = glob.glob(data_path)
data = []
for f1 in files:
    img = cv2.imread(f1)
    img = cv2.resize(img, (28, 28))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img = img.reshape((28, 28))
    data.append(img)
```
