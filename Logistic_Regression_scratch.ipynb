```python
import math
import pandas as pd
import numpy as np
prob=[]
odd_ratio=[]
log_Values=[]
for i in range(11):
    prob.append(i*0.1)
    try:
        odd_ratio.append(prob[i]/(1-prob[i]))
        log_Values.append(math.log(odd_ratio[i]))
    except ZeroDivisionError:
        odd_ratio.append(np.inf)
    except ValueError:
        log_Values.append(np.inf)
```

```python
pd.DataFrame(data=list(zip(prob,odd_ratio,log_Values)),columns=['Prob','Odd Ratio','Log(Odd_Ratio)'])
```
